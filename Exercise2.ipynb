```python
"""
Exercise 2a: Create a circuit ( 𝑈 ) that performs the transformation:

𝑈|00⟩𝑈|01⟩𝑈|10⟩𝑈|11⟩=|01⟩=|10⟩=|11⟩=|00⟩
and is controlled by another qubit. The circuit will act on a 2-qubit target register named 'target',
and be controlled by another single-qubit register named 'control'. You should assign your finished 
circuit to the variable 'cu'.
"""
from qiskit import QuantumCircuit
from qiskit import QuantumRegister, QuantumCircuit
c = QuantumRegister(1, 'control')
t = QuantumRegister(2, 'target')
cu = QuantumCircuit(c, t, name="Controlled 13^x mod 35")


U= QuantumCircuit(3)
U.ccx(0,1,2)
U.cx(0,1)

cu.append(U,[0,1,2])
cu = cu.decompose()
```

```python
"""
Exercise 2b: Create a circuit ( 𝑈2 ) that performs the transformation:

𝑈|00⟩𝑈|01⟩𝑈|10⟩𝑈|11⟩=|10⟩=|11⟩=|00⟩=|01⟩
 
and is controlled by another qubit. The circuit will act on a 2-qubit
target register named 'target', and be controlled by another single-qubit
register named 'control'. You should assign your finished circuit to the variable 'cu2'.
"""
c = QuantumRegister(1, 'control')
t = QuantumRegister(2, 'target')
cu2 = QuantumCircuit(c, t)

# WRITE YOUR CODE BETWEEN THESE LINES - START

U= QuantumCircuit(3)
U.ccx(0,1,2)
U.cx(0,1)

U.ccx(0,1,2)
U.cx(0,1)



cu2.append(U,[0,1,2])
cu2=cu2.decompose()
```

```python
"""
Exercise 2c: Create a circuit ( 𝑈4 ) that performs the transformation:

𝑈|00⟩𝑈|01⟩𝑈|10⟩𝑈|11⟩=|00⟩=|01⟩=|10⟩=|11⟩
 
and is controlled by another qubit. The circuit will act on a 2-qubit target register named 'target', and be controlled by another single-qubit register named 'control'. You should assign your finished circuit to the variable 'cu4'. Hint: The best solution is very simple.

c = QuantumRegister(1, 'control')
t = QuantumRegister(2, 'target')
cu4 = QuantumCircuit(c, t)
"""

U= QuantumCircuit(3)

U.ccx(0,1,2)
U.cx(0,1)

U.ccx(0,1,2)
U.cx(0,1)

U.ccx(0,1,2)
U.cx(0,1)

U.ccx(0,1,2)
U.cx(0,1)

cu4.append(U,[0,1,2])
cu4=cu4.decompose()
```

```python
"""
Exercise 2 final: Now we have controlled  𝑈 ,  𝑈2  and  𝑈4 , we can combine this into a circuit
that carries out the quantum part of Shor’s algorithm.

The initialization part is easy: we need to put the counting register into the state  |+++⟩ 
(which we can do with three H-gates) and we need the target register to be in the state  |1⟩ 
(which we mapped to the computational basis state  |00⟩ , so we don’t need to do anything here). 
We'll do all this for you.

Your task is to create a circuit that carries out the controlled- 𝑈 s, that will be used in-between
the initialization and the inverse quantum Fourier transform. More formally, we want a circuit:

𝐶𝑈𝑐0𝑡𝐶𝑈2𝑐1𝑡𝐶𝑈4𝑐2𝑡
 
Where  𝑐0 ,  𝑐1  and  𝑐2  are the three qubits in the ‘counting’ register,  𝑡  is the ‘target’ register, 
and  𝑈  is as defined in the first part of this exercise. In this notation,  𝐶𝑈𝑎𝑏  means  𝐶𝑈  is
controlled by  𝑎  and acts on  𝑏 . An easy solution to this is to simply combine the circuits cu,
cu2 and cu4 that you created above, but you will most likely find a more efficient circuit that has the same behavior!
"""
cqr = QuantumRegister(3, 'control')
tqr = QuantumRegister(2, 'target')
cux = QuantumCircuit(cqr, tqr)
solutions = [cu, cu2, cu4]
for i in range(3):
    cux = cux.compose(solutions[i], [cqr[i], tqr[0], tqr[1]])
from qiskit import transpile
cux = transpile(cux, basis_gates=['cx','u'])
cux.draw('mpl')
```
